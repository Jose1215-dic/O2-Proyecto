```python
import math

def capacidad_agua_puente(longitud_puente, ancho_puente, altura_agua_maxima):
  """
  Calcula la capacidad de agua de un puente, considerando su geometría y la altura máxima del agua.

  Args:
    longitud_puente: Longitud del puente en metros.
    ancho_puente: Ancho del puente en metros.
    altura_agua_maxima: Altura máxima del agua en metros (por encima del nivel del puente).

  Returns:
    La capacidad de agua del puente en metros cúbicos.
    Devuelve -1 si alguna de las entradas es inválida.
  """


  if longitud_puente <= 0 or ancho_puente <= 0 or altura_agua_maxima < 0:
    print("Error: Las dimensiones del puente deben ser positivas y la altura del agua no puede ser negativa.")
    return -1

  # Calcula el volumen de agua que puede contener el puente
  volumen_agua = longitud_puente * ancho_puente * altura_agua_maxima

  return volumen_agua


# Ejemplo de uso
longitud = 40  # metros
ancho = 12  # metros
altura_maxima = 4  # metros

capacidad = capacidad_agua_puente(longitud, ancho, altura_maxima)

if capacidad != -1:
    print(f"La capacidad de agua del puente es: {capacidad} metros cúbicos")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def plot_bridge_capacity(length, width, height):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')


    x = [0, length, length, 0, 0]
    y = [0, 0, width, width, 0]
    z = [0, 0, 0, 0, 0]

    ax.plot(x, y, z)
    ax.plot([0, length], [0,0], zs=0, zdir='z', label='Longitud')
    ax.plot([0, 0], [0, width], zs=0, zdir='z')


    x_water = np.linspace(0, length, 10)
    y_water = np.linspace(0, width, 10)
    X, Y = np.meshgrid(x_water, y_water)
    Z = np.full(X.shape, height)
    ax.plot_surface(X, Y, Z, alpha=0.5, color='blue', label='Capacidad de Agua')


    ax.set_xlabel('Longitud (m)')
    ax.set_ylabel('Ancho (m)')
    ax.set_zlabel('Altura (m)')
    ax.set_title(f'Capacidad de Agua del Puente Arroyo Seco ({length}m x {width}m x {height}m)')



    ax.legend()

    plt.show()



length = 40
width = 12
height = 1920/(length*width)

plot_bridge_capacity(length, width, height)
```
